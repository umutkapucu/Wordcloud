## WORDCLOUD

Welcome :)
Let's import libraries!!

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image 
from wordcloud import WordCloud, STOPWORDS
```

Then open the file that we will use (ie. the text file of my CV)

```python
uk_cv=open('cv_uk.txt', 'r').read()
print ('File opened!')
```

Need to set stopwords... for further studies, we could add some words in it!!

```python
my_stopwords = set(STOPWORDS)
```

Let's generate the word cloud!

```python
my_cv_wc = WordCloud(width=2000, height=500, background_color='white', max_words=2000, collocations=False, stopwords=my_stopwords)
my_cv_wc.generate(uk_cv)

fig = plt.figure()
fig.set_figwidth(30) # set width
fig.set_figheight(5) # set height

# display the cloud
plt.imshow(my_cv_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Adding new stopwords!!

```python
my_stopwords.add('AŞ')
my_stopwords.add('16h')
my_stopwords.add('re')
## whatever needed... :)

print(my_stopwords)
```

Regenerate wordcloud...

```python
my_cv_wc = WordCloud(width=2000, height=500, background_color='white', max_words=2000, collocations=True, stopwords=my_stopwords)
my_cv_wc.generate(uk_cv)

fig = plt.figure()
fig.set_figwidth(30) # set width
fig.set_figheight(5) # set height

# display the cloud
plt.imshow(my_cv_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Taaa daaa!
Let's save the picture as a png file!

```python
my_cv_wc.to_file('my_cv_wc.png')
```

Let's set a mask... I've used a picture of me :)

```python
wc_mask = np.array(Image.open('uk-nobackground.png'))

plt.imshow(wc_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Yeah! This is me :) Any picture could be used for mask!

```python
my_cv_wc = WordCloud(width=1000, height=1000, background_color='white', max_words=2000, mask=wc_mask, stopwords=my_stopwords)
my_cv_wc.generate(uk_cv)

fig = plt.figure()
fig.set_figwidth(5) # set width
fig.set_figheight(10) # set height

# display the cloud
plt.imshow(my_cv_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Here we have the masked one!

Let's save this new picture...

```python
my_cv_wc.to_file('my_cv_wc_masked.png')
```
